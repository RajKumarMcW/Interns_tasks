```python
from collections import deque
class Path:
    def __init__(self,n,source_x,source_y,desti_x,desti_y):
        self.n=n
        self.source_x=int(source_x)
        self.source_y=int(source_y)
        self.desti_x=int(desti_x)
        self.desti_y=int(desti_y)
    def display(self):
        l=[]
        for i in range(0,self.n):
            ll=[]
            for j in range(0,self.n):
                ll.append("#")
            print(ll)
            l.append(ll)
    def implementation(self):
        q = deque([(self.source_x,self.source_y,1)])
        visit=set((self.source_x,self.source_y))
        dx=[-1, 1, 0, 0, -1, -1, 1, 1]
        dy=[0, 0, -1, 1, -1, 1, -1, 1]
        parent={}
        while q:
            r,c,steps = q.popleft()
            for i in range(0,8):
                x = r + dx[i]
                y = c + dy[i]
                if 0 <= x < self.n and 0 <= y < self.n and (x, y) not in visit:
                    q.append((x, y, steps + 1))
                    visit.add((x, y))
                    parent[(x, y)] = (r, c)
                if (x == self.desti_x and y == self.desti_y):
                    print(steps + 1)
                    current = (x, y)
                    path = [(x, y)]
                    while current != (self.source_x, self.source_y):
                        current = parent[current]
                        path.append(current)

                        print("Path:")
                    for point in reversed(path):
                        print(point)
                    for i in range(0,self.n):
                        for j in range(0,self.n):
                            fi=[]
                            fi.append(i)
                            fi.append(j)
                            fi=tuple(fi)
                            if(fi not in path):
                                print("#",end=" ")
                            else:
                                print("+",end=" ")
                        print()
                    exit()
print("enter the size")
size=int(input())
source_x,source_y=input("Enter source two values: ").split()
desti_x,desti_y=input("Enter destination two values: ").split()
obj=Path(size,source_x,source_y,desti_x,desti_y)
obj.display()
obj.implementation()
```

```python
def random_rotation(image):
    rows, cols, _ = image.shape
    angle = random.randint(-30, 30)
    M = cv2.getRotationMatrix2D((cols / 2, rows / 2), angle, 1)
    return cv2.warpAffine(image, M, (cols, rows))

def horizontal_flip(image):
    return cv2.flip(image, 1)

def vertical_flip(image):
    return cv2.flip(image, 0)
```

```python
data = []
```

```python
len(os.listdir(bicycle_file_path))
```

```python
len(data)
```

```python
class_folders = [bicycle_file_path,motorcycle_file_path, bus_file_path, car_file_path, truck_file_path] 
for ind,paths in enumerate(class_folders):
    label = ind
    size = len(os.listdir(paths))
    for img in os.listdir(paths):
            img_path = os.path.join(paths, img)
            arr = cv2.imread(img_path)
            if arr is None:
                print(f"Error loading image: {img_path}")
            else:
                new_arr = cv2.resize(arr, (150, 150))
                data.append([new_arr, label])
                while(size<6000):
                    rotated_image = random_rotation(arr)
                    horizontal_image = horizontal_flip(arr)
                    vertical_image = vertical_flip(arr)
                    rotated_image = cv2.resize(rotated_image, (150, 150))
                    horizontal_image = cv2.resize(horizontal_image, (150, 150))
                    vertical_image = cv2.resize(vertical_image, (150, 150))
                    data.append([rotated_image, label])
                    data.append([horizontal_image, label])
                    data.append([vertical_image, label])
                    size+=3
```

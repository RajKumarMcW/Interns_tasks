```python
X = df.iloc[:, :-1]
y = df['DEATH_EVENT']
X.shape
```

SMOTE(Synthetic Minority Over-sampling Technique) it is used to identify the minority class on the biased data and oversample it, to make it unbiased.

Based on target "Death_event", it oversampled the entire column.

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=2021, k_neighbors=5)
smote.fit(X, y)
X_smote, y_smote = smote.fit_resample(X, y)
print(X_smote.shape)
sns.countplot(df,x=y_smote)
```

PowerTransformer is used to reduce the skewness in the dataset by taking logarithm on data. It has two method methods.

*   box-cox (It is used only for positive values,not include zero)
*   yeo-johnson (It is used for both positive and negative values)







```python
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method='yeo-johnson')
df_transform = pt.fit_transform(df)
pd.DataFrame(df_transform , columns=df.columns).hist(figsize=(10, 10))
plt.show()
```

MinMaxScaler is used for scaling the columns within a range(0,1)

```python
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
scaled = mm.fit_transform(df_transform)
pd.DataFrame(scaled, columns=df.columns).hist(figsize=(10, 10))
plt.show()
```
